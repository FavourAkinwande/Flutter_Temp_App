# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
#TO DO: Create a function that does a derivative for any function
# Import the sympy library as an alias 'smp'
import sympy as smp

# Define a function 'dy_dx' to calculate the derivative of 'fx'
def dy_dx(fx, x):

  derivative = smp.diff(fx, x)

# Return the calculated Derivative
  return derivative

````

# Task 4: Create separate array of Minimas and maximas

````python
# creating sample array
arr = [4, 3, 2, 8, 4, 0, 5, 6]

arr_minimas = []
arr_maximas = []

for i in range(1, len(arr) - 1):
    if arr[i] < arr[i - 1] and arr[i] < arr[i + 1]:  # Check for local minima
        arr_minimas.append((i, arr[i]))  # Store (index, value)
    elif arr[i] > arr[i - 1] and arr[i] > arr[i + 1]:  # Check for local maxima
        arr_maximas.append((i, arr[i]))  # Store (index, value)

# Printing the results
print(f'The array of minimas {arr_minimas} and the array of maximas is {arr_maximas}')

# Finding global results
Global_maxima = max(arr_maximas, key=lambda x: x[1], default=None)  # (index, value)
Global_minima = min(arr_minimas, key=lambda x: x[1], default=None)  # (index, value)

# Printing global minima and maxima
if Global_maxima:
    print(f'The global maxima is {Global_maxima}')
else:
    print('There are no global maxima.')

if Global_minima:
    print(f'The global minima is {Global_minima}')
else:
    print('There are no global minima.')
````

# Task 5: Finding the Area under the curve

````python
 #This is where you do integration

import sympy as smp
import numpy as np
import matplotlib.pyplot as plt


x = smp.symbols('x')
fx = x**2 + 5*x + 6
fx_deriv = smp.diff(fx, x)

func = smp.lambdify(x, fx)
deriv = smp.lambdify(x, fx_deriv)

# Define range for x values
x_vals = np.linspace(-10, 10, 50)
y_vals = func(x_vals)
y_deriv_vals = deriv(x_vals)

# Plot the function and its derivative
plt.figure(figsize=(9,9))
plt.plot(x_vals, y_vals, label='f(x)', color='red', marker='o', linestyle='dashed')
plt.plot(x_vals, y_deriv_vals, label='deriv_f(x)', color='orange', marker='x')

plt.title('Quadratic Function and Derivative')
plt.xlabel('X value')
plt.ylabel('f(x) value')
plt.grid()
plt.legend()
plt.show()

# Finding the area under the curve using sympy integration
area_under_curve = smp.integrate(fx, (x, -10, 10))  # Integrate from x = -10 to x = 10
print(f'The area under the curve from x = -10 to x = 10 is: {area_under_curve}')
````

````python
import sympy as smp
import matplotlib.pyplot as plt
import numpy as np

# Define the symbolic variable
x = smp.symbols('x')

# Define the function fx with at least 2 minima and 2 maxima
fx = x**4 - 4*x**2

# Define a function 'dy_dx' to calculate the derivative of 'fx'
def dy_dx(fx, var):
    return smp.diff(fx, var)

# Find critical points and classify them as minima or maxima
def find_critical_points(fx, var):
    first_derivative = dy_dx(fx, var)
    critical_points = smp.solve(first_derivative, var)
    second_derivative = dy_dx(first_derivative, var)

    # Classify critical points
    return [
        (point, 'minima' if second_derivative.subs(var, point) > 0 else 'maxima')
        for point in critical_points
    ]

# Convert symbolic expressions to numerical functions for plotting
func = smp.lambdify(x, fx)
fx_deriv = dy_dx(fx, x)
deriv = smp.lambdify(x, fx_deriv)

# Create x values for plotting
x_vals = np.linspace(-3, 3, 15)
y_vals = func(x_vals)
y_deriv_vals = deriv(x_vals)

# Plot the function and its derivative
plt.figure(figsize=(9, 9))
plt.plot(x_vals, y_vals, label='f(x)', color='red', marker='o', linestyle='dashed')
plt.plot(x_vals, y_deriv_vals, label='f\'(x)', color='orange', marker='x')

# Find and mark critical points
critical_points = find_critical_points(fx, x)
for point, extrema_type in critical_points:
    point_val = float(point)
    y_val = func(point_val)
    plt.plot(point_val, y_val, 'go' if extrema_type == 'minima' else 'bo',
             label=f'{extrema_type.capitalize()} at x={point_val:.2f}')

# Add labels and title
plt.title('Function and Derivative with Minima and Maxima')
plt.xlabel('X value')
plt.ylabel('f(x) and f\'(x) value')

plt.grid(True)
plt.legend()
plt.show()
````
